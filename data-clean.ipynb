```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd

books = pd.read_csv(f"{path}/books.csv")
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### 利用热力图来查看缺失值的相关规律

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), ax=ax, cbar=False)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

### 利用spearman相关性矩阵来查看删除这些缺失的标签与rate是否有相关性，保证能够删除缺失值

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="YlGnBu",
                      cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation Matrix")
plt.show()
```

```python
books_complete = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["published_year"].isna()) &
      ~(books["average_rating"].isna())
]
```

```python
books_complete
```

### 查看现有分类合理性

```python
books_complete["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

### 计划利用llm分析描述来进行文本分类，筛选出无效描述

```python
books_complete["words_in_description"] = books_complete["description"].str.split().str.len()
```

```python
books_complete
```

```python
books_complete.loc[books_complete["words_in_description"].between(15, 24), "description"]
```

```python
books_missing_25_words = books_complete[books_complete["words_in_description"] >= 25]
```

```python
books_missing_25_words
```

### 绑定现有的副标题

```python
books_missing_25_words["title_and_subtitle"] = (
    np.where(books_missing_25_words["subtitle"].isna(), books_missing_25_words["title"],
        books_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)
    )
)
```

```python
books_missing_25_words
```

### 绑定isbn和描述，便于后续向量嵌入

```python
books_missing_25_words["tagged_description"] = books_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
books_missing_25_words
```

```python
(
    books_missing_25_words.drop(
        ["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1
    ).to_csv("books_cleaned.csv", index=False)
)
```
